###### DROPPING COLUMNS UNSEFUL IN MY ANALYSIS

```python

#dropping columns with large percentage of missing data
df.drop(columns=["Latitude","Longitude",'Investigation.Type', 'Accident.Number', 'Airport.Code', 'Airport.Name',
       'Registration.Number','Amateur.Built','FAR.Description', 'Schedule','Air.carrier', 'Report.Status','Publication.Date'],inplace=True)
```

```python
#filling in the aircract category with data from make and model 
def fill_nan_aircraft_category(df):
    df['Aircraft.Category'] = df.groupby(['Make'])['Aircraft.Category'].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
    return df
fill_nan_aircraft_category(df)
```

## DATA ANALYSIS

### 1.ANALYSING ACCIDENTS DISTRIBUTION OVER TIME

```python
#Extracting year from date time for plotting purposes
df['Date'] = pd.to_datetime(df['Event.Date'])
df['Year'] = df['Date'].dt.year

#count the number of accidents per year
accidents_per_year = df['Year'].value_counts().sort_index()
accidents_per_year
```

```python
#plotting the line graph to show the trend 
sns.lineplot(x=accidents_per_year.index, y=accidents_per_year.values)
plt.title('Number of Accidents per Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.show()
```

This clearly indicates that due to advancement in technology in the 1980s the making of aerolplanes increased and this led to a sharp increase in the accidents over time they learned about it and have aspired to manufacture more sustainable aircrafts thus the reduction in accidents, other factors might be better tech in tracking the aeroplanes, good training of pilots etc, more research can be done to check what impacts more.

### 1.ANALYSING ACCIDENTS DISTRIBUTION OVER TIME BY AIRCRAFT CATEGORY

```python
#plotting the different aircraft categories

plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Aircraft.Category', order=df['Aircraft.Category'].value_counts().index,color="red")
plt.title('Distribution of Accidents by Aircraft Category')
plt.xlabel('Aircraft Category')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=90)
plt.yscale("log")
plt.show()
```

### 2.ANALYSING ACCIDENTS DISTRIBUTION OVER TIME BY MAKE

```python
#finding the top 10 makes with the most accidents
top_makes = df['Make'].value_counts().nlargest(10).index
top_makes
```

```python
#Plotting the 
sns.countplot(data=df[df['Make'].isin(top_makes)], x='Make', order=top_makes)
plt.title('Top 10 Makes with Most Accidents')
plt.xlabel('Make')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=90)
plt.figure(figsize=(14, 6))
plt.show()
```

The visualisation shows the top 10 most common accidents by make with CESSNA being the highest

### 3.ANALYSING AT WHAT POINT THE ACCIDENT HAPPENNED 

```python
#plotting the bar graph to show the broad phase of flight
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='Broad.phase.of.flight', order=df['Broad.phase.of.flight'].value_counts().index,color="blue")
plt.title('Number of Accidents by Broad Phase of Flight')
plt.xlabel('Number of Accidents')
plt.ylabel('Broad Phase of Flight')
plt.show()
```

This visualisation shows that the occurence  of the accident in the phase of flight in most cases is unknown but for the known cases most of the accidents happen at the airport during landing or take off

### 4.DISTRIBUTION OF OUTCOMES(INJURIES)

```python

# Group by Broad.phase.of.flight and calculate totals
phase_of_flight_data = df.groupby('Broad.phase.of.flight').agg(
    total_incidents=('Event.Id', 'count'),
    total_fatal_injuries=('Total.Fatal.Injuries', 'sum'),
    total_serious_injuries=('Total.Serious.Injuries', 'sum'),
    total_minor_injuries=('Total.Minor.Injuries', 'sum')
).reset_index()

# Plot the data
plt.figure(figsize=(14, 8))

# Total incidents by phase of flight
plt.subplot(2, 1, 1)
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_incidents'], color='purple')
plt.title('Total Incidents by Phase of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Total Incidents')
plt.xticks(rotation=45)
plt.grid(True)

# Total injuries by phase of flight
plt.subplot(2, 1, 2)
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_fatal_injuries'], color='blue', alpha=0.5, label='Fatal Injuries')
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_serious_injuries'], color='yellow', alpha=0.5, label='Serious Injuries')
plt.bar(phase_of_flight_data['Broad.phase.of.flight'], phase_of_flight_data['total_minor_injuries'], color='red', alpha=0.5, label='Minor Injuries')
plt.title('Total Injuries by Phase of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Total Injuries')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()




```
